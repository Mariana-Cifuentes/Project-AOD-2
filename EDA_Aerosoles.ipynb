#### **`Mediciones de AOD (Aerosol Optical Depth) - espesor óptico de aerosoles en diferentes longitudes`**

```python
AOD = df_aerosoles.filter(like='AOD')
AOD = AOD.drop(columns=list(conteo.filter(like='N[AOD').columns))

AOD.head()
```

```python
columnas_AOD = AOD.columns
print(columnas_AOD)
```

```python
# Longitudes de onda qe nos sirven
funcionales = AOD[['AOD_340nm', 'AOD_380nm', 'AOD_400nm', 'AOD_412nm', 'AOD_440nm', 'AOD_443nm', 'AOD_490nm', 'AOD_500nm', 'AOD_510nm', 'AOD_532nm', 'AOD_551nm', 'AOD_555nm', 'AOD_560nm', 'AOD_620nm', 'AOD_667nm', 'AOD_675nm', 'AOD_779nm']]

funcionales = funcionales.replace(-999, pd.NA)
```

```python
funcionales = funcionales.replace(-999, np.nan).apply(pd.to_numeric, errors='coerce')

plt.figure(figsize=(14,6))
funcionales.boxplot(rot=45, grid=True)
plt.title('Distribución de valores AOD por longitud de onda', fontsize=14)
plt.ylabel('AOD')
plt.tight_layout()
plt.show()
```

```python
func = funcionales.replace(-999, np.nan).apply(pd.to_numeric, errors='coerce')

plt.figure(figsize=(14,6))
func.boxplot(rot=45, grid=True)
plt.title('AOD per wavelength (zoom near 0)')
plt.ylabel('AOD')

plt.ylim(-0.01, 0.5)   # muestra claramente los valores < 0
plt.axhline(0, ls='--', lw=1)
plt.tight_layout()
plt.show()
```

#### **`Información General de Columnas de interés`**

```python
cols_finales = [
    "AERONET_Site", "Date(dd:mm:yyyy)", "Day_of_Year",
    "Precipitable_Water(cm)", "440-870_Angstrom_Exponent",
    "Site_Latitude(Degrees)", "Site_Longitude(Degrees)", "Site_Elevation(m)",

    "AOD_340nm","AOD_380nm","AOD_400nm","AOD_412nm","AOD_440nm","AOD_443nm",
    "AOD_490nm","AOD_500nm","AOD_510nm","AOD_532nm","AOD_551nm","AOD_555nm",
    "AOD_560nm","AOD_620nm","AOD_667nm","AOD_675nm","AOD_681nm","AOD_709nm",
    "AOD_779nm","AOD_865nm","AOD_870nm","AOD_1020nm","AOD_1640nm"
]
df_sub = df_aerosoles[cols_finales].copy()
```

```python
df_sub.info()
```

```python
mask_invalid = (df_sub == -999).sum().sort_values(ascending=False)
mask_invalid = mask_invalid[mask_invalid > 0]

ax = mask_invalid.plot(kind='bar', figsize=(12,5), color='tomato')
ax.set_title("Number of values ​​-999 per column")
ax.set_ylabel("Frequency of -999")
ax.yaxis.get_major_formatter().set_scientific(False)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_sub_clean = df_sub.replace(-999, np.nan)
```

```python
aod_cols = [c for c in df_sub_clean.columns if c.startswith("AOD_")]

plt.figure(figsize=(14,6))
sns.boxplot(data=df_sub_clean[aod_cols], color='lightblue')
plt.title("Distribución de valores AOD (sin -999)")
plt.ylabel("AOD Value")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
negativos = (df_sub_clean[aod_cols] < 0).sum().sort_values(ascending=False)
negativos = negativos[negativos > 0]

plt.figure(figsize=(12,5))
negativos.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title("Número de valores negativos por longitud de onda (AOD)")
plt.ylabel("Frecuencia de valores < 0")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df_sub_clean["440-870_Angstrom_Exponent"], bins=40, color='orange', kde=True)
plt.title("Distribución del Ångström Exponent")
plt.xlabel("440–870 Ångström Exponent")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df_sub_clean["Precipitable_Water(cm)"], bins=40, color='skyblue', kde=True)
plt.title("Distribución del Precipitable Water")
plt.xlabel("Precipitable Water (cm)")
plt.tight_layout()
plt.show()
```
